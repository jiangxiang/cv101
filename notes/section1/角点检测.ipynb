# 角点检测

角点是指图像中某个点附近存在两个或多个方向的边缘。

这个点周围的像素可以在至少两个不同的方向上被视为边缘，这种情况被称为角点。

角点通常是物体的特征点，可以用于物体跟踪、三维重建等应用。

opencv提供了两种角点检测方法

* harris角点检测

* shi-tomasi角点检测

## harris角点检测原理

基本思想：**利用局部像素灰度变化检测角点**

他定义角点基于两个假设

1. 周围像素区域至少有两个方向的灰度值变化较大--角点

2. 像素周围的像素区域内只有一个方向的灰度值变化较大--边缘



定义角点的物理意义：





当窗口中没有角点和边缘，因此无论怎么滑动，灰度不会变换

当窗口中存在边缘，因此滑动时在某个方向，灰度不会变化

当窗口存在角点，至少存在两个方向的边缘，因此无论往那个方向滑动，都会有剧烈灰度变化

具体步骤：

    1. 计算每个像素的梯度
    2. 计算自相关矩阵
  
    
    3. 计算响应函数
    
   
    
    4. 进行非极大值抑制，得到角点
    

从底层实现角点检测


````python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("../../dataset/grid.png", 0)
ori_img = img.copy()
plt.imshow(img, cmap='gray')
plt.show()
````

````python
ksize = 5
k = 0.04
````

````python
s_x = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
s_y = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)
````

````python
# 计算自相关矩阵
Ixx = s_x ** 2
Ixy = s_x * s_y
Iyy = s_y ** 2
````

cv2.boxFilter 用于对图像进行Box Filter

使用一个大小为 $k \times k$ 的方形窗口，窗口中的每个像素都赋予相同的权重，这些权重之和为1。

````python
# 计算区域响应
Sxx = cv2.boxFilter(Ixx, -1, (3, 3))
Sxy = cv2.boxFilter(Ixy, -1, (3, 3))
Syy = cv2.boxFilter(Iyy, -1, (3, 3))
````

````python
# 计算R
det = Sxx * Syy - Sxy ** 2
trace = Sxx + Syy
R = det - k * trace ** 2
````

````python
# 非极大值抑制
R_max = np.max(R)
R[R < .01 * R_max] = 0
R = cv2.dilate(R, None)
R_binary = np.uint8(R > 0) * 255
````

````python
# 找出角点
corners = []
for i in range(R_binary.shape[0]):
    for j in range(R_binary.shape[1]):
        if R_binary[i, j] > 0:
            corners.append((j, i))
            img[i-5:i+5, j-5:j+5] = 255
````

````python
plt.imshow(img)
plt.show()
````
