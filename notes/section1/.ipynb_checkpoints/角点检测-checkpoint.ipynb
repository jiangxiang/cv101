# 角点检测

角点是指图像中某个点附近存在两个或多个方向的边缘。

这个点周围的像素可以在至少两个不同的方向上被视为边缘，这种情况被称为角点。

角点通常是物体的特征点，可以用于物体跟踪、三维重建等应用。

opencv提供了两种角点检测方法

* harris角点检测

* shi-tomasi角点检测

## harris角点检测原理

基本思想：**利用局部像素灰度变化检测角点**

他定义角点基于两个假设

1. 周围像素区域至少有两个方向的灰度值变化较大--角点

2. 像素周围的像素区域内只有一个方向的灰度值变化较大--边缘



定义角点的物理意义：





当窗口中没有角点和边缘，因此无论怎么滑动，灰度不会变换

当窗口中存在边缘，因此滑动时在某个方向，灰度不会变化

当窗口存在角点，至少存在两个方向的边缘，因此无论往那个方向滑动，都会有剧烈灰度变化

具体步骤：

    1. 计算每个像素的梯度
    2. 计算自相关矩阵
  
    
    3. 计算响应函数
    
   
    
    4. 进行非极大值抑制，得到角点
    

从底层实现角点检测


````python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("../../dataset/grid.png", 0)
ori_img = img.copy()
plt.imshow(img, cmap='gray')
plt.show()
````

````python
ksize = 5
k = 0.04
````

````python
s_x = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
s_y = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)
````

````python
# 计算自相关矩阵
Ixx = s_x ** 2
Ixy = s_x * s_y
Iyy = s_y ** 2
````

cv2.boxFilter 用于对图像进行Box Filter

使用一个大小为 $k \times k$ 的方形窗口，窗口中的每个像素都赋予相同的权重，这些权重之和为1。

````python
# 计算区域响应
Sxx = cv2.boxFilter(Ixx, -1, (3, 3))
Sxy = cv2.boxFilter(Ixy, -1, (3, 3))
Syy = cv2.boxFilter(Iyy, -1, (3, 3))
````

````python
# 计算R
det = Sxx * Syy - Sxy ** 2
trace = Sxx + Syy
R = det - k * trace ** 2
````

````python
# 非极大值抑制
R_max = np.max(R)
R[R < .01 * R_max] = 0
R = cv2.dilate(R, None)
R_binary = np.uint8(R > 0) * 255
````

````python
# 找出角点
corners = []
for i in range(R_binary.shape[0]):
    for j in range(R_binary.shape[1]):
        if R_binary[i, j] > 0:
            corners.append((j, i))
            img[i-5:i+5, j-5:j+5] = 255
````

````python
plt.imshow(img)
plt.show()
````

### shi-tomasi角点检测

如前所述，harris角点检测采用基于梯度的方式实现

tomasi角点检测也是使用梯度的方式进行检测。

采用梯度检测的方式都具有的共同点就是，采用滑动窗口的方式来检测像素点周围的像素变化

* 如果两个变换大，则是一个角点
* 如果两个方向变化小，则不是一个角点



除了从公式理解变化，也可以从特征值的几何角度理解变化检测。

具体上，harris角点检测的$R$值计算公式如下：


$$R=det(M) - k\cdot tract(M)^2$$

其中，$M$为自相关矩阵

det是计算矩阵行列式的值

trace是计算矩阵的迹

而其中特征值$\lambda_1$, $\lambda_2$表示了在x方向和y方向上的变化大小。

本质上，harris角点检测和tomash角点检测方法，都是基于任意方向上的梯度变化来实现的。

harris角点检测用行列式的值和矩阵的迹来评估特征值，从而实现强角点检测；

不同于harris方法，tomasi方法则认为，**是否为角点，则是通过两个特征值的最小值决定的**

因而，tomasi方法则是通过下列公式

$$R=\min{(\lambda_1, \lambda_2)}$$

也就是说，当$R>\lambda_{min}$，则说明是一个角点。

那么tomasi角点检测算法的底层实现步骤为：

* 高斯滤波
* 对每个像素的k邻域，计算自相关矩阵M

$$M=\begin{bmatrix}
\sum_{u,v}w(u, v)I_x^2 & \sum_{u,v}w(u, v)I_xI_y \\
\sum_{u,v}w(u, v)I_xI_y & \sum_{u,v}w(u, v)I_y^2
\end{bmatrix}$$

* 计算M的特征值$\lambda_1$， $\lambda_2$，并计算$R=\min(\lambda_1, \lambda_2)$

* 当R的值超过阈值，则认为是一个角点

**opencv的api**

**cv2.goodFeaturesToTrack(img, maxCorners, quialityLevel, minDistance, mask, blockSize, useHarrisDetector, k)**

* img：目标图像
* maxCorners：最大角点数
* quialityLevel：角点阈值
* minDistance：角点间最小距离
* mask：检测区域
* blockSize：kernel大小
* useHarrisDetector：是否使用harris检测
* k：harris检测的参数


````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../../dataset/lena.png", 0)
corners = cv2.goodFeaturesToTrack(img, 200, 0.01, 10)
corners = np.intp(corners)

print(corners.shape)
````

````python
for i in corners:
    x, y = i.ravel()
    cv2.circle(img, (x, y), 3, 255, -1)
plt.imshow(img)
plt.show()
````

### 角点检测的应用

**利用角点实现图像配准的思路**

1. 对两张图像进行角点检测
2. 对角点区域进行特征提取，每个角点都可以获得一个固定长度的向量表示特征
3. 对角点特征进行两两匹配
4. 对匹配成功的特征进行筛选约束
5. 进行图像配准
