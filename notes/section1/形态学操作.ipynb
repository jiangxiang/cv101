# 形态学操作

定义：基于形状和结构的理论和方法，目的是改变和处理图像中对象的形状和结构，实现图像的进一步处理

主要内容：
    
    - 膨胀 dialation
    - 腐蚀 erosion
    - 开运算 opening
    - 闭运算 closing
    - 形态学梯度 morphological gradiant
    - 顶帽 top hat
    - 底帽 black hat

注意：形态学操作的对象是二值图像，在进行形态学操作之前，需要确认是否已经将图像转换为二值图像。

### 膨胀

是形态学中的一种基本操作，用于增强图像中的白色区域（前景区域）并缩小黑色区域（背景区域）

其原理是用一个结构元素（也称为核或窗口）在图像上滑动，将核的中心点对应的像素值替换为结构元素中的最大值。

膨胀操作可以用于填充图像中的空洞、去除图像中的小物体以及连接分离的物体等。

例如，我们以上周的边缘检测方法为例，用膨胀方法来强化图像的边缘信息。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像并转为灰度图像
img = cv2.imread('../../dataset/greenscreen.png', cv2.IMREAD_GRAYSCALE)

# 定义拉普拉斯滤波器卷积核
kernel = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])

# kernel = np.array([[0, -1, 0],
#                    [-1, 4, -1],
#                    [0, -1, 0]])

# 应用卷积核
filtered_img = cv2.filter2D(img, -1, kernel)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(filtered_img, cmap='gray')
plt.show()
```

要进行膨胀处理，首先应当转化为二值图像

```python
_, binary_img = cv2.threshold(filtered_img, 10, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.imshow(filtered_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

之后，即可进行膨胀运算。

```python
kernel = np.ones((5, 5), np.uint8)
dilation = cv2.dilate(binary_img, kernel, iterations=1)

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(dilation, cmap='gray')
plt.show()
```

#### 膨胀操作小结：

一方面，膨胀操作能填补细小空隙，即把白色像素附近的黑色变成白色；

另一方面，膨胀操作能放大噪点，即把细小的白色变成黑色

### 腐蚀

那么如何将噪声消除呢？

回答：利用腐蚀操作。

与膨胀操作相反，腐蚀操作则是通过用窗口中的最小值来作为当前像素的值。

也就是说，膨胀操作填补黑点，放大白点；

而腐蚀操作是填补白点，放大黑点。

通常，腐蚀可以将边缘变得更加锐利，用于分离粘连在一起的物体，去除毛刺等等。

例如，我们可以对上图中的膨胀结果进行腐蚀。

```python
kernel = np.ones((5, 5), np.uint8)
eroded = cv2.erode(255 - binary_img, kernel, iterations=1)

plt.subplot(1, 2, 1)
plt.imshow(255 - binary_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(eroded, cmap='gray')
plt.show()
```

显然，腐蚀操作可以去除大量噪点。

但是，腐蚀操作同样会消除掉细弱的边缘

**那么如何在保护细弱的边缘的同时，消除掉噪声呢？**

### 开、闭运算

开运算：先腐蚀再膨胀

闭运算：先膨胀再腐蚀

首先需要明白一个原则：

* 如果在腐蚀操作中，把一个孤立的白点完全去除了，再对其进行膨胀会怎么样？

* 如果在膨胀运算中，把一个孤立的黑点完全去除了，再对其进行腐蚀会怎么样？

那么，遵从膨胀腐蚀的操作原则，如果能够在完全腐蚀掉噪声的同时，边缘还剩余一些微弱的信号，那么我们再对其进行膨胀就不会引起噪声的反噬。

上述就是开运算的基本思路。

同理，闭运算是先进行膨胀，后进行腐蚀。这样可以填补一些空洞，再通过腐蚀保持其原有形状。

那么，我们仍以绿幕图像为例，通过人为为期添加椒盐噪声，来演示开闭运算的性能。

```python
import numpy as np
import cv2


img = cv2.imread("../../dataset/cat.jpeg", 0)
kernel = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])
filtered_img = cv2.filter2D(img, -1, kernel)
_, binary_img = cv2.threshold(filtered_img, 10, 255, cv2.THRESH_BINARY)

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 3, 2)
plt.imshow(filtered_img, cmap='gray')
plt.subplot(1, 3, 3)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(255 - binary_img, cv2.MORPH_OPEN, kernel)

plt.subplot(1, 2, 1)
plt.imshow(255 - binary_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(opening, cmap='gray')
plt.show()
```

```python
kernel = np.ones((21, 21), np.uint8)
closing = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(opening, cmap='gray')
plt.show()
```

```python
new_img = img.copy()
new_img[closing == 0] = 0

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(new_img)
plt.show()
```

#### 小结： 

- 开运算和闭运算虽然集成了两种操作，但是只能采用固定的核大小，因此缺少灵活性

- 通常，可以采用更加极端的操作来实现更加精细的结果。

- 对于绿幕分割，实际上可以采用阈值分割、滤波、以及形态学操作多种技术结合，来获得更加鲁棒的结果

### 形态学梯度

思考问题：

    - 给定原始图像$I$
    - 对原始图像进行膨胀$I_d$
    - 对原始图像进行腐蚀$I_e$
    - 令$I_d - I_e$是什么？


上述过程被称为**形态学梯度**

形态学梯度中包含的主要信息为：

    - 边缘信息
    - 轮廓信息
    - 高频噪声信息
    
形态学梯度主要的内容是将图像中平滑的区域去除，只保留边缘

以下是对lena图像进行形态学梯度的运算结果。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../../dataset/lena.png', 0)
_, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
dilation = cv2.dilate(binary_img, kernel, iterations=1)
eroded = cv2.erode(binary_img, kernel, iterations=1)

plt.subplot(1, 2, 1)
plt.imshow(dilation, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(eroded, cmap='gray')
plt.show()
```

```python
gradient = np.abs(dilation - eroded)
plt.imshow(gradient, cmap='gray')
plt.show()
```

实际上，可以通过不同等级像素值的二值化方法，来得到不同等级的形态学梯度。

这些梯度之间再进行融合，即可得到更加精细的边缘结果。

### 顶帽运算和底帽运算

* 顶帽运算：原始图像减去开运算结果

* 底帽运算：原始图像检测闭运算结果

那么顶帽运算和底帽运算会对图像产生什么物理意义？

```python
import cv2

img = cv2.imread('../../dataset/0.jpg', 0)
img = cv2.resize(img, [128, 128])
noise = np.zeros_like(img)
cv2.randu(noise, 0, 255)
salt = noise > 250
pepper = noise < 5
img[salt] = 255
img[pepper] = 0

_, binary_img = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

```python
tophat = cv2.morphologyEx(binary_img, cv2.MORPH_TOPHAT, kernel).copy()
new_img = np.abs(tophat - binary_img)
kernel = np.ones((3, 3), np.uint8)
new_img = cv2.morphologyEx(new_img, cv2.MORPH_CLOSE, kernel).copy()
plt.subplot(1, 2, 1)
plt.imshow(tophat, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(new_img, cmap='gray')
plt.show()
```

显然，顶帽检测用于亮区检测，去除大的亮斑点，来获取小的亮点和线条。


那么底帽运算与其相反，我们可以通过底帽运算来检测出黑点的位置，用于缺陷检测等场景

例如，去除大的暗点，突出小暗点和暗线条等。

```python
blackhat = cv2.morphologyEx(binary_img, cv2.MORPH_BLACKHAT, kernel).copy()
new_img = np.abs(blackhat - binary_img)
kernel = np.ones((3, 3), np.uint8)
plt.subplot(1, 2, 1)
plt.imshow(blackhat, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(new_img, cmap='gray')
plt.show()
```
